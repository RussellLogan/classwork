```python
import pandas as pd
def eda(df: pd.DataFrame) -> None:
    """
    Do stat eval on dataframe
    
    Args"
        df(pd.DataFrame): Dataframe to evaluatate
    Returns:
        None
    """
    num_cols = []
    cat_cols = []
    for col in df.columns:
        if pd.api.types.is_numeric_dtype(df[col]):
            num_cols.append(col)
        else:
            cat_cols.append(col)
    print(f"{'='*5}DF SHAPE: {df.shape}{'='*5}")
    print(f"{'*'*3}Numeric Columns{'*'*3}")
    for col in num_cols:
        print(f"{col}")
        max = df[col].max()
        min = df[col].min()
        print(f"\tMin = {min}")
        print(f"\tMax = {max}")
        print(f"\tRange = {max - min}")
```

```python
eda(df)
```
